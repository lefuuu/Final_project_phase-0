### Импортируем библиотеки

```python
import pandas as pd
import numpy as np
import re
```

### Загрузка данных

```python
pd.set_option('display.max_columns', None)                          # устанавливаем отображение всех столбцов
df = pd.read_csv('../data/_data.csv', index_col='Unnamed: 0')       # Загружаем файл
df.head(5)                                                          # Выводим первые 5 строк для проверки
```

### Выводим список столбцов с количеством уникальных значений

```python
feature_names = df.columns.tolist() 
for column in feature_names: 
    print(column, '=' ,df[column].nunique())
```

### Удаляем столбцы, не имеющие смысловой нагрузки для модели

```python
df.drop(['Тип', 'Телефоны', 'Дополнительно' , 'Название ЖК', 'Серия дома', 'Ссылка на объявление'], axis = 1, inplace=True)
# столбец "Тип" нам не подходит, ввиду отсутствия в нем вариантов (1 уникальное значение)
# остальные столбцы, на наш взгляд, содержат информацию, не влияющую на работу модели машинного обучения
```

### Обрабатываем колонку "Количество комнат"

```python
def extract_digits(text):                                                                                   # вспомогательная функция, вытягивающая из строки числа
    digits = re.findall(r'\d+', str(text))
    if len(digits) == 1:
        return int(digits[0])
    return sum(map(int, digits))


df['Количество комнат'] = df['Количество комнат'].apply(extract_digits)                                     # перезаполняем колонку "Количество комнат"
df['Количество комнат'] = df['Количество комнат'].fillna(df['Площадь комнат, м2'].apply(extract_digits))    # заполняем пустые значения, суммируя площадь комнат (если указано)
df.drop(['Площадь комнат, м2'], axis=1, inplace=True)                                                       # удаляем ненужный больше столбец "Площадь комнат"
```

### Обрабатываем колонку "Метро"

```python
def time_from_metro(text):                                          # вспомогательная функция, вытягивающая из строки число, идущее перед строкой "мин пешком"
    match = re.search(r'\((\d+)\s*мин\s*пешком\)', str(text))
    if match:
        return int(match.group(1))
    else:
        return None

df['Метро'] = df['Метро'].apply(time_from_metro)                    # перезаполняем колонку "Метро"
df['Метро'] = df['Метро'].fillna(0)                                 # заполняем пустые значения 0, считаем, что 0 - расстояние до метро не известно
# К сожалению, мы не знаем, как модель обрабатывает входные данные, поэтому здесь мы сделали на основе наших предположений
# мы считали, что высчитывать расстояние до метро у пропущенных значений не совсем корректно, ввиду того, что это должно быть какое-то константное значение, 
# а не рассчитанное. В то же время, оно никак не связано с другими характеристиками, поэтому внесенное нами значение (рассчитанное по любой из формул) может
# привести к искажению результата работы модели
# В то же время, мы предположили, что модель может быть настроена таким образом, что любая цифра, отличная от 0, может считаться указанным расстоянием до метро,
# в то время, как 0 означает, что значение не указано. Основываясь на этом предположении мы и реализовали данную графу.
```

### Обрабатываем колонку "Адрес"

```python
df = df[df['Адрес'].str.contains("Москва")].reset_index(drop = True)    # исключаем из датасета записи, в которых адрес не равен значению Москва
df.drop('Адрес', axis=1, inplace=True)                                  # удаляем колонку "Адрес" за ненадобностью
```

### Обрабатываем колонку "Площадь, м2"

```python
def extract_first_digit(text):                                          # вспомогательная функция, вытягивающая из строки первое число
    values = re.search(r'\d+', str(text))
    return int(values.group())
df['Площадь, м2'] = df['Площадь, м2'].apply(extract_first_digit)        # перезаполняем колонку "Площадь, м2"
```

### Обрабатываем колонку "Дом", создаем 2 новых колонки "Этаж" и "Высота дома"

```python
df['Этаж']=df['Дом'].str.split(',').str[0].str.split('/').str[0]                        # создаем колонку "Этаж" на основе данных в графе "Дом"
df['Высота дома']=df['Дом'].str.split(',').str[0].str.split('/').str[1]                 # создаем колонку "Высота дома" на основе данных в графе "Дом"

df.drop(['Дом'], axis=1, inplace=True)                                                  # удаляем колонку "Дом" за ненадобностью
```

### Обрабатываем колонку "Парковка"

```python
df['Парковка'] = df['Парковка'].fillna('0').replace({'подземная': '1', 'наземная': '2', 'открытая': '3', 'многоуровневая': '4', 'на крыше': '5'})
# меняем данные в графе "Парковка", основываясь на следующие характеристики:
# 0 - парковка отсутствует
# 1 - подземная парковка
# 2 - наземная парковка
# 3 - открытая парковка
# 4 - многоуровневая парковка
# 5 - парковка на крыше
```

### Обрабатываем колонку "Цена", создаем колонку "Коммунальные услуги включены"

```python
def public_utilities(text):                                                 # вспомогательная функция, находящая вхождение подстроки "Коммунальные услуги включены" в строку
    match = re.search('Коммунальные услуги включены', str(text))
    return 1 if match else 0

df['Коммунальные услуги включены'] = df['Цена'].apply(public_utilities)     # перезаполняем колонку "Коммунальные услуги включены"
df['Цена']=df['Цена'].apply(extract_first_digit)                            # перезаполняем колонку "Цена"
```

### Создаем колонку "Цена за квадратный метр", чтобы с её помощью заполнить пустые поля в поле количество комнат

```python
df['Цена за квадратный метр'] = df['Цена'] / df['Площадь, м2']                              # создаем новую колонку 'Цена за квадратный метр'
df['Цена за квадратный метр'] = df['Цена за квадратный метр'].round(2)                      # округляем значения в этой колонке до 2-х знаков после запятой

res = df.groupby(['Количество комнат'])['Цена за квадратный метр'].median()                 # находим среднее значение цены за квадратный метр с группировкой по 
del res[0]                                                                                  # количеству комнат
def quanttity_room(square_price):                                                           # # вспомогательная функция, определяющая количество комнат в зависимости
    for i in range(1, len(res)):                                                            # от стоимости квадратного метра
        if res[i] <= square_price <= res[i + 1]:
            return i if (square_price - res[i]) < (res[i + 1] - square_price) else i + 1
        elif square_price < res[1]:
            return 1
        elif square_price > res[len(res)]:
            return len(res)

df['Количество комнат'] = df.apply(lambda row: quanttity_room(row['Цена за квадратный метр']) if row['Количество комнат'] == 0 else row['Количество комнат'], axis=1)
# Заполняем количество комнат в позициях, где у нас был 0
# Можно было считать количество комнат в зависимости от площади квартиры, или даже в зависимости от площади и цены за квадратный метр,
# но к сожалению, у нас было ограничение по времени, и опять же, мы можем только предполагать, как работает модель. 
# В итоге решили остановиться на этом варианте
```

### Обрабатываем колонку "Ремонт"

```python
def extract_renovation(description):                                                    # вспомогательная функция, находящая вхождение подстрок из списка в исходную строку
    patterns = ['Дизайнерский', 'Евроремонт', 'Косметический', 'Без ремонта']
    
    for remont_type in patterns:
        if re.search(remont_type, str(description), re.IGNORECASE):
            return remont_type
    return 'Не указан'

df['Ремонт'] = df.apply(lambda row: extract_renovation(row['Описание']) if pd.isna(row['Ремонт']) else row['Ремонт'], axis=1)
# перезаполняем колонку "Ремонт", заполняем ее данными из колонки "Описание" (при наличии), если данные в колонке "Ремонт" отсутствуют

df['Ремонт'] = df['Ремонт'].replace({'Не указан':'0', 'Без ремонта':'1', 'Косметический':'2', 'Евроремонт':'3', 'Дизайнерский':'4'})
# заменяем данные в колонке "Ремонт" на числовой признак согласно следующему шаблону:
# 0 - Не указан
# 1 - Без ремонта
# 2 - Косметический ремонт
# 3 - Евроремонт
# 4 - :'3', 'Дизайнерский ремонт

df.drop(['Описание'], axis=1, inplace=True)         # удаляем колонку "Описание" за ненадобностью
```

### Обрабатываем колонку "Балкон"

```python
def extract_balcony(text):                                                      # вспомогательная функция, находящая вхождение указанных подстрок в исходную строку
    if re.search('Балкон', str(text)) and re.search('Лоджия', str(text)):
        return 3
    elif re.search('Лоджия', str(text)):
        return 2
    elif re.search('Балкон', str(text)):
        return 1
    else:
        return 0
    
df['Балкон']=df['Балкон'].fillna(0).apply(extract_balcony)                      
# заменяем данные в колонке "Балкон" на числовой признак согласно следующему шаблону:
# 0 - не указано
# 1 - есть только балкон
# 2 - есть только лоджия
# 3 - есть балкон и лоджия
```

### Обрабатываем колонку "Окна"

```python
df['Окна'] = df['Окна'].fillna('0').replace({'На улицу':'1', 'Во двор':'2', 'На улицу и двор':'3'})
# заменяем данные в колонке "Окна" на числовой признак согласно следующему шаблону:
# 0 - не указано
# 1 - на улицу
# 2 - во двор
# 3 - на улицу и двор
```

### Обрабатываем колонку "Санузел"

```python
def extract_bathroom(text):                                                             # вспомогательная функция, находящая вхождение указанных подстрок в исходную строку
    if re.search('Раздельный', str(text)) and re.search('Совмещенный', str(text)):
        return 3
    elif re.search('Раздельный', str(text)):
        return 2
    elif re.search('Совмещенный', str(text)):
        return 1
    else:
        return 0
    
df['Санузел']=df['Санузел'].fillna(0).apply(extract_bathroom)
# заменяем данные в колонке "Санузел" на числовой признак согласно следующему шаблону:
# 0 - не указано
# 1 - совмещенный санузел
# 2 - раздельный санзел
# 3 - совмещенный и раздельный санузел
```

### Обрабатываем колонку "Можно с детьми/животными"

```python
def replace_children_pets(text):                                                                # вспомогательная функция, находящая вхождение указанных подстрок в исходную строку
    patterns = ['Можно с детьми, Можно с животными', 'Можно с детьми', 'Можно с животными']
    for allow in patterns:
        if re.search(allow, str(text)):
            return allow
    return 'Не указан'

df['Можно с детьми/животными'] = df['Можно с детьми/животными'].fillna('0').replace({'Можно с детьми':'1', 'Можно с животными':'2', 'Можно с детьми, Можно с животными':'3'})
# заменяем данные в колонке "Можно с детьми/животными" на числовой признак согласно следующему шаблону:
# 0 - не указано
# 1 - можно с детьмиМетро
# 2 - можно с животными
# 3 - можно с детьми и животными
```

### Обрабатываем колонку "Высота потолков, м"

```python
median_value = df['Высота потолков, м'].median()                            # считаем среднее значение по колонке
df['Высота потолков, м'] = df['Высота потолков, м'].fillna(median_value)    # заполняем пропуски средним значением (посчитали нецелесообразным применение более сложных формул)
```

### Обрабатываем колонку "Лифт"

```python
def extract_elevator(text):                                                     # вспомогательная функция, находящая вхождение указанных подстрок в исходную строку
    if re.search('Пасс', str(text)) and re.search('Груз', str(text)):
        return 3
    elif re.search('Груз', str(text)):
        return 2
    elif re.search('Пасс', str(text)):
        return 1
    else:
        return 0
    
df['Лифт']=df['Лифт'].fillna(0).apply(extract_elevator)
# заменяем данные в колонке "Лифт" на числовой признак согласно следующему шаблону:
# 0 - не указано
# 1 - пассажирский
# 2 - грузовой
# 3 - пассажирский и грузовой
```

### Обрабатываем колонку "Мусоропровод"

```python
df['Мусоропровод'] = df['Мусоропровод'].fillna('0').replace({'Нет': '1', 'Да': '2'})
# меняем содержимое в графе "Мусоропровод" на числовой признак: 0 - не указано, 1 - мусоропровода нет, 2 - мусоропровод есть
```

### Переносим колонку "Цена" в конец датасета

```python
price_col = df['Цена']                                  # сохраняем исходную колонку "Цена"
df.drop(['Цена'], axis=1, inplace=True)                 # удаляем её из датафрейма
df.insert(len(df.columns), 'Цена', price_col)           # вставляем сохранённую колонку "Цена" в конец датафрейма
```
