```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('data/_data.csv', index_col='Unnamed: 0') 
df.head()

```

```python
df.info()
```

```python
df.nunique() 
# как видно из серии, колонки Тип, Парковка, Ремонт, Окна, Можно с детьми/животными и Мусоропровод 
#идеально подходят для категориальных признаков
```

```python
# Визуализация пропущенных значений (столбчатая диаграмма)
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
plt.figure(figsize=(10, 6))
missing_values.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Количество пропущенных значений по столбцам', fontsize=16)
plt.xlabel('Столбцы', fontsize=12)
plt.ylabel('Количество пропущенных значений', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('graphs/missing_values_bar.png')  # Сохранение графика
plt.show()
```

```python
df['Цена'] = df['Цена'].astype('str')
df['Цена,руб/м'] = df['Цена'].str.extract(r'(\d+\.?\d*)').astype(float) # новый столбец с числовыми значениями цены аренды за  месяц
df['Количество комнат'] = pd.to_numeric(df['Количество комнат'], errors='coerce') # преобразование столбца к числовому формату, потом Nan заполним
```

```python
# График распределение цен
plt.figure(figsize=(10, 6))
plt.hist(df['Цена,руб/м'], bins = 40, edgecolor='black')
plt.minorticks_on()
plt.title('Распределение цен')
plt.xlabel('Цена за месяц')
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='x')
plt.xlim(0, 400000)
plt.ylabel('Количество квартир')
plt.savefig('graphs/prices_count.png')  # Сохранение графика
```

```python
# Построение графика зависимости цены от количества комнат
plt.figure(figsize=(10, 6))
plt.scatter(df['Количество комнат'], df['Цена,руб/м'])
plt.title('Зависимость цены от количества комнат')
plt.xlabel('Количество комнат')
plt.ylabel('Цена,руб/м')
plt.ylim(-100000,3100000)
plt.grid(True)
plt.savefig('graphs/prices_by_rooms.png')  # Сохранение графика
plt.show()
```

```python
# Построение графика зависимости расстояния от метро от цены
plt.figure(figsize=(10, 6))
plt.scatter(df['Метро'].str.extract(r'(\d+)').astype(float),  df['Цена,руб/м'], alpha=0.5) # по оси х извлекаем числовое значение "n минут"
plt.title('Зависимость расстояния до метро от цены')
plt.xlabel('Расстояние до метро (мин)')
plt.xlim(-10,100)
plt.ylabel('Цена')
plt.grid(True)
plt.savefig('graphs/subw_dist_by_price.png')  # Сохранение графика
plt.show()
```

```python
df['Площадь, м2'] = df['Площадь, м2'].str.split('/').str[0].astype(float) # Извлечение первой цифры из столбца "Площадь, м2"
#Связь цены и общей площади и ремонта
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Площадь, м2'], y= df['Цена,руб/м'], hue=df['Ремонт'])
plt.title('Связь цены и общей площади и ремонта')
plt.xlabel('Общая площадь, м2')
plt.xlim(right=350)
plt.ylabel('Цена, руб/мес')
plt.ylim(top=2000000)
plt.ticklabel_format(style='plain', axis='y');
plt.grid()
plt.savefig('graphs/price_area_furn.png')  # Сохранение графика
```

```python
#Связь цены и типа ремонта
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ремонт', y='Цена,руб/м', data=df)
plt.title('Связь цены и типа ремонта')
plt.xlabel('Тип ремонта')
plt.ylabel('Цена, руб/мес')
plt.ylim(0, 310000)
plt.grid()
plt.savefig('graphs/price_furn.png')  # Сохранение графика
plt.show()
```

```python
#Связь цены и наличия парковки
plt.figure(figsize=(10, 6))
sns.boxplot(x='Парковка', y='Цена,руб/м', data=df)
plt.title('Связь цены и наличия парковки')
plt.xlabel('Наличие парковки')
plt.ylabel('Цена, руб/мес')
plt.ylim(0, 310000)
plt.grid()
plt.savefig('graphs/price_parking.png')  # Сохранение графика
plt.show()
```
